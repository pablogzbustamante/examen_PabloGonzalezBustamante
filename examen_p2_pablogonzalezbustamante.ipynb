```python
class Producto:
    def __init__(self, id, nombre, precio):
        self.__idMD = id
        self.__nombreMD = nombre
        self.__precioMD = float(precio)

    def getId(self):
        return self.__idMD

    def getNombre(self):
        return self.__nombreMD

    def getPrecio(self):
        return self.__precioMD

    def aplicar_descuento(self, porcentajeDes):
        descuento = self.__precioMD * (porcentajeDes / 100)
        self.__precioMD -= descuento

    def calcular_envio(self):
        pass 


class ProductoFisico(Producto):
    def __init__(self, id, nombre, precio, peso):
        super().__init__(id, nombre, precio)
        self.__pesoMD = float(peso)

    def getPeso(self):
        return self.__pesoMD

    def calcularEnvio(self):
        return self.__pesoMD * 10


class ProductoDigital(Producto):
    def __init__(self, id, nombre, precio, tamanho_archivo):
        super().__init__(id, nombre, precio)
        self.__tamanho_archivoMD = float(tamanho_archivo)

    def getTamanhoArchivo(self):
        return self.__tamanho_archivoMD

    def calcularEnvio(self):
        return 0  


```

```python
from pathlib import Path
import csv

#C:\Users\pablo\OneDrive\Documentos\Universidad\Ingeniería en Inteligencia de Datos y Ciberseguridad\Semestre 2\Programacion OO\Examen 2 - Practicas etc\productos.csv

inventario = []

ruta = Path(r"C:\Users\pablo\OneDrive\Documentos\Universidad\Ingeniería en Inteligencia de Datos y Ciberseguridad\Semestre 2\Programacion OO\Examen 2 - Practicas etc\productos.csv")
if ruta.exists():
    with ruta.open("r", encoding="utf-8") as archivo:
        archivoAleer = csv.reader(archivo)
        next(archivoAleer) 
        for fila in archivoAleer:
            if len(fila) >= 5:
                iDMD = fila[0]
                nombre = fila[1]
                precio = float(fila[2])
                tipo = fila[3]
                extra = float(fila[4])

            if tipo == "Fisico":
                producto = ProductoFisico(iDMD, nombre, precio, extra)
            elif tipo == "Digital":
                producto = ProductoDigital(iDMD, nombre, precio, extra)
            else:
                continue  
            inventario.append(producto)
else:
    print("Nadota, no hay archivo")
```

```python
print("Simulacion de compra de productos")

print("Elija un producto:")
for i, producto in enumerate(inventario):
    print(f"Indice: {i+1}. {producto.getNombre()}, con un costo de {producto.getPrecio():.2f} ")


nProductos = input("Escribe los indices de los productos q quiere comprar separados con una coma")
indices = nProductos.split(",")
carritodecompras = []

for i in indices: 
    i = i.strip()
    if i.isdigit():
        indiceCorrecto = int(i) - 1
        if 0 <= indiceCorrecto < len(inventario):
            carritodecompras.append(inventario[indiceCorrecto])
        else:
            print(f"Indice {i} no es correcto")
    else:
        print(f"{i} no es un numero")


if len(carritodecompras) >= 1:
    carritodecompras[0].aplicar_descuento(25) 
if len(carritodecompras) >= 2:
    carritodecompras[-1].aplicar_descuento(25)

costoTotal = 0
envioTotal = 0


if(len(carritodecompras) > 1):
    for i in range(len(carritodecompras)):
        costoTotal += carritodecompras[i].getPrecio()

for i in range(len(carritodecompras)):
    envioTotal += carritodecompras[i].calcularEnvio()


precioYenvioTotal = costoTotal + envioTotal


rutaSalida = Path("reciboExamen.txt")
with rutaSalida.open("w", encoding="utf-8") as archivo:
    archivo.write("====== RECIBO DE COMPRA ======\n\n")
    for producto in carritodecompras:
        
        archivo.write(f"ID del Producto: {producto.getId()}\n")
        archivo.write(f"Nombre del Producto: {producto.getNombre()}\n")
        archivo.write(f"Precio del Producto: ${producto.getPrecio():.2f}\n")
        
        if isinstance(producto, ProductoFisico):
            archivo.write(f"Tipo del Producto: Fisico\nPeso del Producto: {producto.getPeso()} kg\n")
        
        elif isinstance(producto, ProductoDigital):
            archivo.write(f"Tipo del Producto: Digital\nTamaño del Producto: {producto.getTamanhoArchivo()} MB\n")
       
        archivo.write(f"Costo del Envio del Producto: ${producto.calcularEnvio():.2f}\n")
        archivo.write("-------------------------------\n")
    
    archivo.write(f"Compra final:\n")
    archivo.write(f"\nCosto total de productos: ${costoTotal:.2f}")
    archivo.write(f"\nCosto total de envío: ${envioTotal:.2f}")
    archivo.write(f"\nTOTAL A PAGAR: ${precioYenvioTotal:.2f}")

print(f"\nEl recibo esta generado en: {rutaSalida.resolve()}")

```
